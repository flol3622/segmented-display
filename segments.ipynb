```python
# number of changes


def changes(data):
    changes = 0
    for i in range(len(data) - 1):
        if data[i] != data[i + 1]:
            changes += 1
    return changes


fig, axs = plt.subplots(4, 4, figsize=(20, 12))
axs = axs.ravel()
for i, segm in enumerate(data_matrix.T):
    axs[i].plot(segm)
    axs[i].set_title(f"Segment {i+1}, Changes: {changes(segm)}")
    axs[i].set_xlabel("Character")
    axs[i].set_ylabel("State")

# add space between subplots vertically
plt.tight_layout()
plt.show()
```

```python
def total_changes(data):
    total_changes = 0
    for segm in data:
        total_changes += changes(segm)
    return total_changes


total_changes(data_matrix)
```

```python
import numpy as np


def total_changes(matrix):
    # Calculate the number of changes needed between successive rows
    return np.sum(np.abs(np.diff(matrix, axis=0)))


def simulated_annealing(data, initial_temp, cooling_rate, max_iter):
    current_solution = data.copy()
    best_solution = current_solution.copy()
    best_cost = total_changes(best_solution)
    current_temp = initial_temp

    for i in range(max_iter):
        for j in range(len(data)):
            # Swap two rows to create a new candidate solution
            idx1, idx2 = np.random.randint(0, len(data), size=2)
            current_solution[[idx1, idx2]] = current_solution[[idx2, idx1]]

            current_cost = total_changes(current_solution)
            cost_diff = current_cost - best_cost

            # If the new solution is better, accept it
            if cost_diff < 0 or np.exp(-cost_diff / current_temp) > np.random.rand():
                if current_cost < best_cost:
                    best_cost = current_cost
                    best_solution = current_solution.copy()
            else:
                # Revert swap
                current_solution[[idx1, idx2]] = current_solution[[idx2, idx1]]

        # Cool down
        current_temp *= cooling_rate

    return best_cost, best_solution


# Define your matrix here
data_matrix = np.array(
    [[int(bit) for bit in code.replace(" ", "")] for code in segm_13.values()]
)
initial_temp = 10.0
cooling_rate = 0.99
max_iter = 10000  # Adjust based on the computational power available

min_changes, min_perm = simulated_annealing(
    data_matrix, initial_temp, cooling_rate, max_iter
)
print("Minimum changes:", min_changes)
```
